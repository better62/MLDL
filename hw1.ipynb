```python
def find_val_error(y_pred, y_val):
  y_val = y_val['default_Yes'].tolist()
  y_pred = y_pred.tolist()
  y_pred_label = []

  for y in y_pred:
    if y > 0.5:
      y_pred_label.append(1)
    else:
      y_pred_label.append(0)

  mis = 0
  for i in range(len(y_val)):
    if y_val[i] != y_pred_label[i]:
      mis += 1

  val_error = mis/len(y_val)

  return val_error
```

```python
X = default[['income', 'balance']]
y = default[['default_Yes']]
```

```python
from sklearn.utils import shuffle

def k_fold(k, X, y):

  X, y =  shuffle(X, y, random_state = 50)

  slicing = len(X)//k
  val_error_list = []

  for i in range(k):
    start = slicing*i
    end = slicing*(i+1)
    if i == k-1:
      end = len(X)

    X_val = X[start:end]
    y_val = y[start:end]
    X_train = X.drop(X.index[start:end])
    y_train = y.drop(y.index[start:end])

    logit = sm.Logit(y_train, sm.add_constant(X_train)).fit()
    y_pred = logit.predict(sm.add_constant(X_val))
    val_error_tmp = find_val_error(y_pred, y_val)
    val_error_list.append(val_error_tmp)

  return val_error_list
```

```python
val_error_list = k_fold(5, X, y)
```

```python
val_error_list
```

```python
sum(val_error_list)/len(val_error_list)
```

##2-(d)

```python
X = default[['income', 'balance', 'student_Yes']]
y = default[['default_Yes']]
```

```python
def k_fold_2(k, X, y):

  X, y =  shuffle(X, y, random_state = 50)

  slicing = len(X)//k
  val_error_list = []

  for i in range(k):
    start = slicing*i
    end = slicing*(i+1)
    if i == k-1:
      end = len(X)

    X_val = X[start:end]
    y_val = y[start:end]
    X_train = X.drop(X.index[start:end])
    y_train = y.drop(y.index[start:end])

    logit = sm.Logit(y_train, sm.add_constant(X_train)).fit()
    y_pred = logit.predict(sm.add_constant(X_val))

    # mse 구하기
    y_val = y_val['default_Yes'].tolist()
    tmp_error = []
    for idx, value in enumerate(y_pred):
      tmp = (y_val[idx]-value)*(y_val[idx]-value)/len(y_val)
      tmp_error.append(tmp)
    val_error_list.append(sum(tmp_error)/len(tmp_error))
  return val_error_list
```

```python
# regression
val_error_list_regression = k_fold_2(5, X, y)
```

```python
sum(val_error_list_regression)/len(val_error_list_regression)
```

```python
# classification

val_error_list = k_fold(5, X, y)
```
